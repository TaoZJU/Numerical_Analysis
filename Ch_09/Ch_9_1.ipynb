## 9.1 Random Numbers

### Q. 1

```python
m = 2**31 - 1
a = 7**5

seed = 1
```

```python
n = 10**6
x = seed
total = 0

for _ in range(n):
    x = a * x % m
    y = a * x % m
    z = a * y % m
    
    ux = x / m
    uy = y / m
    uz = z / m
    
    if (ux - 1/3)**2 + (uy - 1/3)**2 + (uz - 1/2)**2 < 0.04**2:
        total += 1

print(total / n)
print(4/3 * pi * (0.04)**3)
```

```python
# (b)
area = 0
n = 10**6

for i in range(n):
    x = random()
    area += -2*x**2 + 2*x

print(area / n)
```

```python
# (c)
area = 0
n = 10**6

x_list = []
y_list = []

for i in range(n):
    x = random()
    y = random()
    
    if x**2 - x + 1/2 < y and -x**2 + x + 1/2 > y:
        x_list.append(x)
        y_list.append(y)
        area += 1

fig = plt.figure(figsize=(5, 5))
plt.scatter(x_list[:10**4], y_list[:10**4], s=0.1)
plt.axis([0, 1, 0, 1])
plt.show()
        
print(area / n)
```

### Q. 5

```python
# (a)
area = 0
n = 10**4

x_list = []
y_list = []

for i in range(n):
    x = 2 * random() - 1
    y = 2 * random() - 1

    if 13*x**2 + 34*x*y + 25*y**2 <= 1:
        x_list.append(x)
        y_list.append(y)
        area += 1

fig = plt.figure(figsize=(5, 5))
plt.scatter(x_list[:10**4], y_list[:10**4], s=0.1)
plt.axis([-1, 1, -1, 1])
plt.show()
        
print((area / n) * 4)
print(pi / 6)
```

```python
# (b)
area = 0
n = 10**4

x_list = []
y_list = []

for i in range(n):
    x = random()
    y = random()

    if 40*x**2 + 25*y**2 + y + 9/4 <= 52*x*y + 14*x:
        x_list.append(x)
        y_list.append(y)
        area += 1

fig = plt.figure(figsize=(5, 5))
plt.scatter(x_list[:10**4], y_list[:10**4], s=0.1)
plt.axis([0, 1, 0, 1])
plt.show()
        
print(area / n)
print(pi / 18)
```

### Q. 6

```python
area = 0
n = 10**6

x_list = []
y_list = []
z_list = []

for i in range(n):
    x = random()
    y = random()
    z = random()

    if 2 + 4*x**2 + 4*z**2 + y**2 <= 4*x + 4*z + y:
        x_list.append(x)
        y_list.append(y)
        z_list.append(z)
        area += 1

print(area / n)
print(pi / 24)
```

```python
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x_list[:10**4], y_list[:10**4], z_list[:10**4], s=0.1)

ax.set_xlim3d(0, 1)
ax.set_ylim3d(0, 1)
ax.set_zlim3d(0, 1)
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_zlabel('z label')

plt.show()
```

```python
# (b)
area = 0
n = 10**6

x_list = []
y_list = []

for i in range(n):
    x = random()
    y = random()

    if x**2 < y and y < sqrt(x):
        x_list.append(x)
        y_list.append(y)
        area += x*y

fig = plt.figure(figsize=(5, 5))
plt.scatter(x_list[:10**4], y_list[:10**4], s=0.1)
plt.axis([0, 1, 0, 1])
plt.show()
        
print(area / n)
print(1/12)
```
