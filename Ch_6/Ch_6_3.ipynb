## 6.3 Systems of Ordinary Differential Equations

```python
def Euler_method(f, a, b, h, y_0):
    n = int((b - a)/h)
    d = len(y_0)
    
    t = np.linspace(a, b, n+1)
    w = np.zeros((d, n+1))
    w[:, 0] = y_0
    
    for i in range(n):
        w[:, i+1] = w[:, i] + h*(f(t[i], w[:, i]))
        
    return t, w
```

```python
def Trapezoid_method(f, a, b, h, y_0):
    n = int((b - a)/h)
    d = len(y_0)
    
    t = np.linspace(a, b, n+1)
    w = np.zeros((d, n+1))
    w[:, 0] = y_0
    
    for i in range(n):
        w[:, i+1] = w[:, i] + (f(t[i], w[:, i]) + f(t[i] + h, w[:, i] + h*f(t[i], w[:, i])))*(h/2)
        
    return t, w
```

```python
def f(t, y):
    y1, y2 = y
    return np.array([y1 + y2, -y1 + y2])

def exact_sol(t):
    return np.array([exp(t)*cos(t), -exp(t)*sin(t)])
```

```python
h = 0.1
t, y = Euler_method(f, a, b, h, y_0)

plt.plot(t, y[0], label="Euler's Method $y_1$")
plt.plot(t, y[1], label="Euler's Method $y_2$")
plt.plot(t, exact_sol(t)[0], label="Exact $y_1$")
plt.plot(t, exact_sol(t)[1], label="Exact $y_2$")
plt.legend()

plt.show()
```

```python
global_truc_error = abs(exact_sol(t)[:, -1] - y[:, -1])
print("Error at t = 1:", global_truc_error)
```

```python
def f(t, y):
    x1, y1, v1x, v1y, x2, y2, v2x, v2y = y
    
    m1 = 0.3
    m2 = 0.03
    g = 1
    
    P1 = (x1, y1, m1)
    P2 = (x2, y2, m2)
    
    def force(P1, P2, position):
        x1, y1, m1 = P1
        x2, y2, m2 = P2
        if position == 'x':
            return g*m2*(x2 - x1) / sqrt((x2 - x1)**2 + (y2 - y1)**2)**3
        elif position == 'y':
            return g*m2*(y2 - y1) / sqrt((x2 - x1)**2 + (y2 - y1)**2)**3
    
    f1 = [v1x, v1y, force(P1, P2, 'x'), force(P1, P2, 'y')]
    f2 = [v2x, v2y, force(P2, P1, 'x'), force(P2, P1, 'y')]
    
    return np.array(f1 + f2)    
```

```python
a, b = 0, 100
h = 0.005

y_01 = (2, 2) + (0.2, -0.2)
y_02 = (0, 0) + (-0.01, 0.01)
y_0 = y_01 + y_02
```

```python
t, y = Trapezoid_method(f, a, b, h, y_0)
x1, y1, _, _, x2, y2, _, _ = y

plt.plot(x1, y1, label="M1")
plt.plot(x2, y2, label="M2")
plt.title("Trajectory of M1 & M2")
plt.legend()

plt.show()
```
