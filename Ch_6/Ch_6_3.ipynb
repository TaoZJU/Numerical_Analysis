## 6.3 Systems of Ordinary Differential Equations

```python
def Euler_method(f, a, b, h, y_0):
    n = int((b - a)/h)
    d = len(y_0)
    
    t = np.linspace(a, b, n+1)
    w = np.zeros((d, n+1))
    w[:, 0] = y_0
    
    for i in range(n):
        w[:, i+1] = w[:, i] + h*(f(t[i], w[:, i]))
        
    return t, w
```

```python
def Trapezoid_method(f, a, b, h, y_0):
    n = int((b - a)/h)
    d = len(y_0)
    
    t = np.linspace(a, b, n+1)
    w = np.zeros((d, n+1))
    w[:, 0] = y_0
    
    for i in range(n):
        w[:, i+1] = w[:, i] + (f(t[i], w[:, i]) + f(t[i] + h, w[:, i] + h*f(t[i], w[:, i])))*(h/2)
        
    return t, w
```

### Q. 1

```python
# (a)
a, b = 0, 1
y_0 = (1, 0)
```

```python
def f(t, y):
    y1, y2 = y
    return np.array([y1 + y2, -y1 + y2])

def exact_sol(t):
    return np.array([exp(t)*cos(t), -exp(t)*sin(t)])
```

```python
h = 0.1
t, y = Euler_method(f, a, b, h, y_0)

plt.plot(t, y[0], label="Euler's Method $y_1$")
plt.plot(t, y[1], label="Euler's Method $y_2$")
plt.plot(t, exact_sol(t)[0], label="Exact $y_1$")
plt.plot(t, exact_sol(t)[1], label="Exact $y_2$")
plt.legend()

plt.show()
```

```python
global_truc_error = abs(exact_sol(t)[:, -1] - y[:, -1])
print("Error at t = 1:", global_truc_error)
```

```python
h = 0.01
t, y = Euler_method(f, a, b, h, y_0)

plt.plot(t, y[0], label="Euler's Method $y_1$")
plt.plot(t, y[1], label="Euler's Method $y_2$")
plt.plot(t, exact_sol(t)[0], label="Exact $y_1$")
plt.plot(t, exact_sol(t)[1], label="Exact $y_2$")
plt.legend()

plt.show()
```

```python
global_truc_error = abs(exact_sol(t)[:, -1] - y[:, -1])
print("Error at t = 1:", global_truc_error)
```

The reduction in error is consistent with the order of Euler's Method.

### Q. 4

```python
# (a)
d = 1
A = 10

def f(t, y):
    y1, y2 = y
    g = 9.81
    l = 1
    return np.array([y2, -g/l*sin(y1) - d*y2 + A*sin(t)])
```

```python
a, b = 0, 100
h = 0.005
y_0 = (pi, 0)
```

```python
t, y = Trapezoid_method(f, a, b, h, y_0)

plt.plot(t, y[0], label="Angle")
plt.plot(t, y[1], label="Angular velocity")
plt.legend()

plt.show()
```

The pendulum wind around the fixed point three times, stops when the pendulum stands vertical. After that, it repeatedly does the same action in the opposite direction.

```python
y_0 = (2*pi, 0)
```

```python
t, y = Trapezoid_method(f, a, b, h, y_0)

plt.plot(t, y[0], label="Angle")
plt.plot(t, y[1], label="Angular velocity")
plt.legend()

plt.show()
```

The pendulum's moving probably ends up at the same periodic trajectory regardless of the initial condition.

```python
# (b)
a, b = 0, 100
h = 0.1
y_0 = (pi/2, 0)
```

```python
t, y = Trapezoid_method(f, a, b, h, y_0)

plt.plot(t, y[0], label="Angle")
plt.plot(t, y[1], label="Angular velocity")
plt.legend()

plt.show()
```

The terms $-\frac{g}{l}\sin(y_1)$ and $A \cdot \sin(t)$ consistantly amplify the irregular behavior of $\sin(h)$ where $h \in \mathbb{Q}$. Moreover, when $h=0.1$, it is big enough to ignore the constantly changing value of $\sin$ function.<br>
However, since $\sin(x)$ is a periodic function, the integrate motion shows a repeated action.
